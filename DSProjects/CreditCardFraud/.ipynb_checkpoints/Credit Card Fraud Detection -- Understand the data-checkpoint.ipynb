# Introduction

Greetings! Before we starting to use any model on the Fraud Detection, firstly we need to understand the data we have. In this notebook, we will conduct the exploratory analysis of the Credit Card Fraud Detection dataset.

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### 1. Loading the dataset

Loading the dataset and checking out the first 10 rows of data. 

```python
fd_data = pd.read_csv(r"/Users/Vivian/Documents/PyProjects/XW-Datashelf/DSProjects/CreditCardFraud/data/creditcard.csv")
fd_data.head(10)
```

#### 2. Checking Target column:  `Class`

```python
# count class
class_count = pd.value_counts(fd_data["Class"])
class_count.plot("bar")
plt.title("Non-Fraud vs. Fraud")
plt.xlabel("Class")
plt.ylabel("Frequency")
```

Comments on the result... 1. the data has been desensitized. 2. maybe a result from PCA. 3. Data is unbalanced

#### 3. Checking Independent Varibales

```python
# 30 variables
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(15, 15))
counter = 0
for i in range(4):
    for j in range(4):
        ax = axes[i][j]
        # Plot when we have data
        if counter < len(fd_data.columns):
            ax.hist(fd_data[fd_data.columns[counter]], bins=20, color='blue', alpha=0.5, label='{}'.format(fd_data.columns[counter]))
            ax.set_xlabel(fd_data.columns[counter])
            ax.set_ylabel('PDF')
            counter +=1
```
